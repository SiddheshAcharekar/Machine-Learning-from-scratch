```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, multivariate_normal
import random
```

```python
def simpleRandomGaussian(mu, sigma, n_samples, width=4):
    output = []
    while len(output) != n_samples:
        # generate a random number within specific (4 by default) standard deviations of mean
        x = np.random.choice(np.arange(mu-width*sigma, mu+width*sigma, 0.1))
        # generate a random number between 0 and 1
        y = np.random.sample(1)[0]
        # get pdf of x
        pdf_x = norm.pdf(x, mu, sigma)

        if y < pdf_x:
            output.append(x)
    return output
gaussian_samples = simpleRandomGaussian(50, 2, 300)
plt.title("Histogram of simple gaussian sampling")
plt.hist(gaussian_samples)
plt.show()
```

```python
def simpleRandomUniform(min, max, n_samples):
    output = []
    while len(output) != n_samples:
        # generate a random number within specific (4 by default) standard deviations of mean
        x = np.random.choice(np.arange(min, max, 0.1))
        # generate a random number between 0 and 1
        y = np.random.sample(1)[0]
        # get cdf of x
        cdf_x = x-min/(max-min)
        if y < cdf_x:
            output.append(x)
    return output
uniform_samples = simpleRandomUniform(0,200, 30000)
plt.title("Histogram of simple uniform sampling")
plt.hist(uniform_samples)
plt.show()
```

```python
mu_t = np.array([0,0])
cov_t = np.array([[1,3/5], [3/5, 1]])
def simpleRandom2DGaussian(mu_2d, sigma_2d, n_samples, width = 4):
    output = np.zeros((n_samples,2))
    i = 0
    while i != n_samples:

#         x = np.random.choice(np.arange(mu_2d[0]-width*sigma_2d[0,1], mu_2d[0]+width*sigma_2d[0,1], 0.1),2)
        x_1 = np.random.choice(np.arange(mu_2d[0]-width*sigma_2d[0][0], mu_2d[0]+width*sigma_2d[0][0], 0.1))
        x_2 = np.random.choice(np.arange(mu_2d[1]-width*sigma_2d[1][1], mu_2d[1]+width*sigma_2d[1][1], 0.1))
        
        y = np.random.sample(1)[0]
        
        pdf_x = multivariate_normal.pdf([x_1,x_2], mean = mu_2d, cov = sigma_2d)
        if y < pdf_x:
            output[i,:] = np.array([x_1,x_2])
            i += 1
            
    return output

gaussian2d_samples = simpleRandom2DGaussian(mu_t, cov_t, 1000)
plt.title("Histogram of 2-D Gaussian sampling")
# plt.hist(gaussian2d_samples)
plt.scatter(gaussian2d_samples[:,0], gaussian2d_samples[:,1])
plt.show()
```

```python
def sampling_stevens():
    def discretePDF(data):
        freq = np.bincount(data)
        relative_freq = freq/np.float32(data.shape[0])
        nz = relative_freq.nonzero()[0]
#         print('ZIP:',list(zip(nz, relative_freq[nz])))
        return np.asarray(list(zip(nz, relative_freq[nz])))
    
    def simpleRandomWithoutReplacement(d, n_samples):
        data = np.asarray(d)
        def get_bucket(bucket_uniform_cumulative, y):
            j = 0
            for b in bucket_uniform_cumulative:
                if(b > y):
                    break
                j += 1
            return j

        def random_element(sorted_bucket, bucket_num):
            i = 0
            for sb in sorted_bucket:
                if i == bucket_num:
                    result = random.choice(sorted_bucket[i])
                    break
                i += 1
            return result

        output = []

        # get pdf of each dataset
        pdf = discretePDF(data)
        print('Sorted pdf check',np.argsort(-pdf[:,1]))
        sorted_pdf = pdf[np.argsort(-pdf[:,1])]

        plt.title("Probability distributions")
        plt.bar(pdf[:,0], pdf[:,1])
        plt.show()

        plt.title("Sorted probability distributions")
        plt.bar(range(1,sorted_pdf[:,0].shape[0]+1), sorted_pdf[:,1])
        plt.xticks(range(1, sorted_pdf[:,0].shape[0] + 1), sorted_pdf[:,0])
        plt.show()

        j = 0
        bucket = []
        record = []
        for i in sorted_pdf[:,1]:
            record.append(i)
            if(j % n_samples == n_samples-1):
                bucket.append(record)
                record = []
            j += 1
        if(len(record) > 0):
            bucket.append(record)

        record = []
        sorted_bucket = []
        j = 0
        for i in sorted_pdf[:,0]:
            record.append(i)
            if(j % n_samples == n_samples-1):
                sorted_bucket.append(record)
                record = []
            j += 1
        if(len(record) > 0):
            sorted_bucket.append(record)

        bucket_uniform = [np.sum(b) for b in bucket]

        bucket_uniform_cumulative = np.cumsum(bucket_uniform)

        # generate n_sample random numbers between 0 and 1
        y = np.random.sample(n_samples)
        assert len(y) == n_samples
        for yy in y:
            bucket_num = get_bucket(bucket_uniform_cumulative, yy)
            proceed = True
            while(proceed):
                element = random_element(sorted_bucket, bucket_num)
                if element in d:
                    output.append(element)
                    d.remove(np.int32(element))
                    proceed = False
        return output
    sample_input = np.random.randint(1, 500, 300)
    print('Input samples:', sample_input)
    samples_without_replacement = simpleRandomWithoutReplacement(list(sample_input), 20)
    plt.title("Histogram of samples")
    plt.hist(samples_without_replacement)
    plt.show()
    print('Output samples:', samples_without_replacement)
```
